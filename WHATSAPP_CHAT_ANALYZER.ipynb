```python
# extract date and message in different columns

pattern = r'\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\u202f?(?:AM|PM)\s-\s'

all_messages = []
all_dates = []

for cleaned_data in raw:
    dates = re.findall(pattern, cleaned_data, re.IGNORECASE)
    all_dates.extend(dates)
    
    messages = re.split(pattern, cleaned_data, flags=re.IGNORECASE)[1:]
    all_messages.extend(messages)

print(len(all_messages))
print(len(all_dates))
```

```python
def get_time_and_date(string):
    string = string.split(',')
    date,time = string[0],string[1]
    time = time.split('-')
    time = time[0].strip()
    
    return date+" "+time
```

```python
df = pd.DataFrame({"user_messages":all_messages,
                  "date":all_dates})

df["date"] = df["date"].apply(lambda text:get_time_and_date(text))
```

```python
users = []
messages = []

for message in df['user_messages']:
    entry = re.split('([\w\W]+?):\s',message)
    if entry[1:]:
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append("Group Notification")
        messages.append(entry[0])
        
df['User'] = users
df['message'] = messages

def getstring(text):
    return text.split('\n')[0]

df["message"] = df["message"].apply(lambda text:getstring(text))

df.head()
```

```python
def mask_phone_number(text):
    pattern = r'\+91 \d{5} \d{5}'
    masked_text = re.sub(pattern, lambda m: f'+91 xxxxxx{m.group(0)[-2:]}', text)
    return masked_text
```

```python
df['User'] = df['User'].apply(mask_phone_number)
```

```python
message_counts = df['User'].value_counts()
```

```python
top_10_users = message_counts.nlargest(10)
bottom_10_users = message_counts.nsmallest(10)
```

```python
plt.figure(figsize=(10, 6))
top_10_users.plot(kind='bar')
plt.title('Number of Messages per User - top 10 users')
plt.xlabel('User')
plt.ylabel('Number of Messages')
plt.xticks(rotation=45)
plt.show()

print("---------------------------------------------------------------------------------------------------------------")

plt.figure(figsize=(10, 6))
bottom_10_users.plot(kind='bar')
plt.title('Number of Messages per User - bottom 10 users')
plt.xlabel('User')
plt.ylabel('Number of Messages')
plt.xticks(rotation=45)
plt.show()
```
